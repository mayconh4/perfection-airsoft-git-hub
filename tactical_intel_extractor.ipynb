```python
import google.generativeai as genai
import json

# Configuração da API
# Substitua pelo seu Google API Key
GOOGLE_API_KEY = "SUA_CHAVE_AQUI"
genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel('gemini-1.5-pro')
```

## 🎯 Geração de Inteligência
Defina os parâmetros abaixo para gerar um novo dossiê tático.

```python
def generate_tactical_lore(name, category):
    prompt = f"""
    Atue como um Especialista em Táticas de Infantaria e Airsoft de Elite.
    Gere um dossiê tático para: {name} ({category}).
    
    O texto deve ser em Português Brasileiro, com tom sério, profissional e imersivo (estilo militar/HUD).
    
    Retorne EXATAMENTE no formato JSON abaixo:
    {{
        "fullDescription": "Uma descrição técnica de 3 a 5 parágrafos explicando a origem, objetivo e dinâmica tática.",
        "tacticalTips": [
            "Dica prática 1 de posicionamento ou equipamento",
            "Dica técnica 2 de comunicação ou liderança",
            "Dica estratégica 3 de como vencer a missão"
        ]
    }}
    """
    
    response = model.generate_content(prompt)
    return response.text

# Exemplo de Uso:
# intel = generate_tactical_lore("Milsim: Real Action", "Modalidade")
# print(intel)
```
